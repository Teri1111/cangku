```python
def barw(ax):
    for p in ax.patches:
        val = p.get_width()
        x = p.get_x() + p.get_width()
        y = p.get_y() + p.get_height()
        ax.annotate(round(val, 2),(x, y))


plt.figure(figsize = (15, 30))
ax0 = sns.countplot(y = labels_dataframe['label'], order = labels_dataframe['label'].value_counts().index)
barw(ax0)
plt.show()
           
```
